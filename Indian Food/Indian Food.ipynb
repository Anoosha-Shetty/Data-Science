```python
# Function to replace the ingredients with a more common name and find the frequency of each 
# ingredient

def Ingrep(inp):
    
    word_cnt = {}
    inp = inp.str.strip().str.title()  # Convert the string to Title case

    for i in range(len(inp)):
        str_lst = []
        for j in inp[i].split(','):
            if map_dict.get(j.strip()):
                tmp = map_dict[j.strip()]
            else:
                tmp = '*****To be Mapped*****'
            
            if tmp in word_cnt:
                word_cnt[tmp] = word_cnt[tmp] + 1
            else:
                word_cnt[tmp] = 1
                
            str_lst.append(tmp)
            
        str_lst.sort()

        inp[i] = ', '.join(str_lst)
    
    return inp, word_cnt
```

### Approach for filling the null and missing values
1. In the column 'Flavor_Profile', replace the value -1 with the value 'Mixed'
2. Compare the columns 'Region' and 'State'
 - If Region is provided but the corresponding State is -1 then randomly assign the state value from one of the values present for the states under that region.
 - If Region is not provided but the corresponding State is provided then assign the region value accordingly.
 - If both are missing then assign a random region and assign a random state corresponding to that region

```python
print('No. of records where "Region" is provided but "State" is not provided:',
    len(df[(df['Region']!='-1') & (df['State'] == '-1')]))
print('No. of records where "Region" is not provided but "State" is provided:',
    len(df[(df['Region']=='-1') & (df['State'] != '-1')]))
print('No. of records where "Region" and "State" are not provided:',
    len(df[(df['Region']!='-1') & (df['State'] != '-1')]))
print('\nRegion "null":\n',df[(df['Region'].isnull()) ])
```

```python
# Assign the valid unique values to variables
Regn = list(df[(df['Region']!= '-1') & (df['Region'].notnull())].Region.unique())
Southern_States = list(df[(df['Region']== 'South') & (df['State'] != '-1')].State.unique())
Northern_States = list(df[(df['Region']== 'North') & (df['State'] != '-1')].State.unique())
Eastern_States = list(df[(df['Region']== 'East') & (df['State'] != '-1')].State.unique())
Western_States = list(df[(df['Region']== 'West') & (df['State'] != '-1')].State.unique())
Central_States = list(df[(df['Region']== 'Central') & (df['State'] != '-1')].State.unique())
NorthEast_States = list(df[(df['Region']== 'North East') & (df['State'] != '-1')].State.unique())

print('Region:',Regn)
print('South States:',Southern_States)
print('North States:', Northern_States)
print('East States:',Eastern_States)
print('West States:',Western_States)
print('Central States:',Central_States)
print('North East States:',NorthEast_States)
```

```python
import random
random.seed(100)


# Column 'Flavor_Profile' - replace the value -1 with the value 'Mixed'
df['Flavor_Profile'].replace('-1','Mixed',inplace=True)

# Fill null value in Region

df['Region'].fillna(random.choice(Regn),inplace=True) 


# Randomly assign a region to the rows where Region = -1
df['Region'].replace('-1',random.choice(Regn),inplace=True)

# Based on the region value, randomly assign a state value
df.loc[(df['Region'] == 'South') & (df['State'] == '-1'),'State']=random.choice(Southern_States)
df.loc[(df['Region'] == 'North') & (df['State'] == '-1'),'State']=random.choice(Northern_States)
df.loc[(df['Region'] == 'East') & (df['State'] == '-1'),'State']=random.choice(Eastern_States)
df.loc[(df['Region'] == 'West') & (df['State'] == '-1'),'State']=random.choice(Western_States)
df.loc[(df['Region'] == 'North East') & (df['State'] == '-1'),'State']=random.choice(NorthEast_States)
df.loc[(df['Region'] == 'Central') & (df['State'] == '-1'),'State']=random.choice(Central_States)
```
