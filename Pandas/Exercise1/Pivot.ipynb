```python
#Pivot tables
import pandas as pd
birth = pd.read_csv(r'D:\GitHub\Data-Science\Pandas\Exercise1\US_births_1994-2003_CDC_NCHS.csv')
birth.head(2)
```

```python
month_births = birth.pivot_table(index='month',values='births',aggfunc=sum)
month_births
```

```python
avg_births_mnth = birth.pivot_table(index='month',values='births')
avg_births_mnth
```

```python
#Find the sum total of births monthwise, datewise, daywise
birth.pivot_table(index = ["month", "date_of_month"], columns = "day_of_week", values = "births", aggfunc=sum)
```

```python
birth.pivot_table(index = ["month", "date_of_month", "day_of_week"], values = "births", aggfunc=sum)
```
